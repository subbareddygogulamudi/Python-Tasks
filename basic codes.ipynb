25.Write a program to find the largest among three numbers entered by the user.


```python
a,b,c=map(int,input("Enter the values:").split())
if a>b and a>c:
    print("A is largest")
elif b>a and b>c:
    print("B is largest")
else:
    print("C is largest")
```
